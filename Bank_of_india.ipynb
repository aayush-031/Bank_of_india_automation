```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
def scrapweb(url):
    
    driver.get(url)
    obj=driver.find_elements_by_class_name('tenderbox')
    obj_dict={}
    dfs=pd.DataFrame()
    for t in obj:
        title=t.find_element_by_tag_name('h2').text
    #     print(title)
        lis=t.find_elements_by_tag_name('li')
        url=t.find_element_by_tag_name("a").get_attribute("href")
    #     print(lis[0].text)
    #     print(lis[1].text)
    #     print(lis[2])
        Reference_no=lis[0].text.replace('Reference No.','')
        Date=lis[1].text.replace('Date','')
        obj_dict[Reference_no]={
            'Title' : title,
            'Date' : Date,
            'Reference No' : Reference_no,
            'Detail' : url
        }
        df=pd.DataFrame(obj_dict).T
        dfs=dfs.append(df)
    #     print(Date)
    #     print(url)
    return(dfs)
   
```

```python
driver = webdriver.Chrome()
url1='https://www.bankofindia.co.in/Dynamic/Tender?Type=3'
df_saleNotice= scrapweb(url1)
url2='https://www.bankofindia.co.in/Dynamic/Tender?Type=4'
df_tender=scrapweb(url2)
driver.close()
```
